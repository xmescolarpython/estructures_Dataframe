```python
#Exercici 1
#Descarrega el data set Airlines Delay: Airline on-time statistics and delay causes i carrega’l a un pandas Dataframe. 
#Explora les dades que conté, i queda’t únicament amb les columnes que consideris rellevants.

import os
import pandas as pd

airlines_df = pd.read_csv('Python/DelayedFlights.csv')

airlines_df[:10] # Mostra 10 registres del data set Airlines Delay
airlines_df[["Year","DepTime","CRSDepTime","ArrTime","CRSArrTime"]]
```

```python
#Exercici 2.a
import statistics as stat

#Fes un informe complet del data set:.
airlines_df.describe()
#Resumeix estadísticament les columnes d’interès
airlines_df[["ActualElapsedTime","CRSElapsedTime","AirTime","TaxiIn","TaxiOut"]].describe()
```

```python
#Exercici 2.a

# Dades estructurals
print(airlines_df[['ActualElapsedTime','CRSElapsedTime','AirTime','TaxiIn','TaxiOut']].info())

# Dades númeriques
airlines_df[['ActualElapsedTime','CRSElapsedTime','AirTime','TaxiIn','TaxiOut']].describe()

# Mitja de les columnes seleccionades
print("Mean")
print(airlines_df[['ActualElapsedTime','CRSElapsedTime','AirTime','TaxiIn','TaxiOut']].mean())
print('\n')

# Mitjana de les columnes seleccionades
print("Median")
print(airlines_df[['ActualElapsedTime','CRSElapsedTime','AirTime','TaxiIn','TaxiOut']].median())
print('\n')

# Valor de Moda a les columnes seleccionades
print("Mode")
print(airlines_df[['ActualElapsedTime','CRSElapsedTime','AirTime','TaxiIn','TaxiOut']].mode())
print('\n')

# Valor de Varianza a les columnes seleccionades
print("Variance")
print(airlines_df[['ActualElapsedTime','CRSElapsedTime','AirTime','TaxiIn','TaxiOut']].var())
print('\n')
```

```python
#Exercici 2.b
#Troba quantes dades faltants hi ha per columna
pd.DataFrame(airlines_df.isnull().sum()) 
```

```python
#Exercici 2.c
#Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)
airlines_df = airlines_df.assign(MeanSpeed = 0)
airlines_df = airlines_df.assign(Late = False)

airlines_df[["Year","DepTime","CRSDepTime","ArrTime","CRSArrTime","MeanSpeed","Late"]]     
            
```

```python
#Exercici 2.d
#Taula de les aerolínies amb més endarreriments acumulats
airlines_df[['UniqueCarrier','ArrDelay','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']].groupby(['UniqueCarrier']).count().sort_values(by=['ArrDelay','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay'],ascending=False).head()
```

```python
#Exercici 2.e.1
#Quins són els vols més llargs? 
airlines_df[["UniqueCarrier","FlightNum","Distance"]].sort_values(by='Distance',axis=0,ascending=False,inplace = False).head()
```

```python
#Exercici 2.e.2
#I els més endarrerits?
airlines_df[["UniqueCarrier","FlightNum","ArrDelay","CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]].sort_values(by=['ArrDelay','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay'],axis=0,ascending=False,inplace = False).head()
```

```python
#Exercici 3
#Exporta el data set net i amb les noves columnes a Excel.
airlines_df.to_csv('Python/airlines.csv') 
airlines_df[["UniqueCarrier","FlightNum","ArrDelay","CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]].sort_values(by=['ArrDelay','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay'],axis=0,ascending=False,inplace = False).head().to_excel('Python/airlines2.xls')
```
